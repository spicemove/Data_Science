```python
import pandas as pd

message = pd.read_csv('SMSSpamCollection.txt', sep='\t', header=None, names=['label', 'message'])

message
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(message)):
    review=re.sub('[^a-zA-z]',' ',message['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2500, binary=True)
X = cv.fit_transform(corpus).toarray()
```

```python
X[1]
```

```python
y = pd.get_dummies(message['label'])

y = y.iloc[:,1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_test, y_pred)

score
```

```python
clsf_rpt = classification_report(y_pred, y_test)

print(clsf_rpt)
```

```python
# Create TFIDF

from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(max_features=2500)
x = tv.fit_transform(corpus).toarray()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,Y_train,Y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

spam_detect_model_tfidf = MultinomialNB().fit(x_train, Y_train)
```

```python
Y_pred = spam_detect_model_tfidf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score_tfidf = accuracy_score(Y_test,Y_pred)

score_tfidf
```

```python
clsf_rpt_tfidf = classification_report(Y_pred, Y_test)

print(clsf_rpt_tfidf)
```

```python
# word2vec

import gensim
import re
from nltk.corpus import stopwords
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(message)):
    review=re.sub('[^a-zA-z]',' ',message['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
        
```

```python
model = gensim.models.Word2Vec(words,window=5, min_count=2)
```
