```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
df = pd.read_csv('zomato.csv', encoding='latin1')

df.head()
```

```python
from matplotlib.pyplot import viridis


sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')

df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on = 'Country Code', how='left')

final_df.head(2)
```

```python
country_names = final_df.Country.value_counts().index

country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3], labels=country_names[:3],autopct='%1.2f%%')
```

```python
rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})

rating.head()
```

```python
import matplotlib


matplotlib.rcParams['figure.figsize']=(12, 6)
sns.barplot(data=rating, x='Aggregate rating', y='Rating count', hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(data=rating,x='Rating color',palette=['blue','red','orange','yellow','green','green'])
```
